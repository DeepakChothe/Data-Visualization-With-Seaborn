# Data Visualization Using Seaborn 

```python
# Lets import the required library
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
tip=sns.load_dataset("tips")
tip.head()
```

# Line Plot


```python
#Lets Work on tips data set
sns.lineplot(x="size",y='total_bill',data=tip)
```

```python
sns.lineplot(x="size",y='total_bill',data=tip,hue="sex")
```

```python
sns.lineplot(x="size",y='total_bill',data=tip,hue="sex",style='sex')
```

```python
sns.lineplot(x="size",y='total_bill',data=tip,hue='sex',style='sex',markers=True)
plt.xlabel("Size",fontsize=15)
plt.ylabel('Total Bill',fontsize=15)
plt.title('Line Plot',fontsize=20)
plt.show()
```

```python
sns.set()
plt.figure(figsize=(10,7)) #  setting width(16) and hight(9) of the figure
sns.lineplot(x="size",y='total_bill',data=tip,hue='sex',style='sex',markers=True,size="sex")
plt.xlabel("Size",fontsize=15)
plt.ylabel('Total Bill',fontsize=15)
plt.title('Line Plot',fontsize=20)
plt.show()
```
